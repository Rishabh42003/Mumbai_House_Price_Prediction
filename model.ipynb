```python
def outlier_remover_sq(df):
  df_output = pd.DataFrame()
  for key,subdf in df.groupby('Location'):
    m = np.mean(subdf.Price_per_sqfeet)
    st = np.std(subdf.Price_per_sqfeet)

    gen_df = subdf[(subdf.Price_per_sqfeet > (m - st)) & (subdf.Price_per_sqfeet <= (m + st))]
    df_output = pd.concat([df_output,gen_df], ignore_index=True)
  return df_output
house = outlier_remover_sq(house)
house.describe()
```

```python
def outlier_remover_bedroom(df):
  exclude_indices=np.array([])
  for Location,Location_df in df.groupby('Location'):
    Bedroom_stats = {}
    for Bedroom,Bedroom_df in Location_df.groupby('Bedroom'):
      Bedroom_stats[Bedroom] = {
          'mean': np.mean(Bedroom_df.Price_per_sqfeet),
          'std': np.std(Bedroom_df.Price_per_sqfeet),
          'count': Bedroom_df.shape[0]      
      }
    for Bedroom,Bedroom_df in Location_df.groupby('Bedroom'):
      stats=Bedroom_stats.get(Bedroom-1)
      if stats and stats['count'] >= 4:
        exclude_indices=np.append(exclude_indices,Bedroom_df[Bedroom_df.Price_per_sqfeet<(stats['mean'])].index.values)
  return df.drop(exclude_indices,axis='index')
    
```

```python
x=house.drop(columns=['Price'])
y=house['Price']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, train_size = 0.7 , random_state = 2529)
```

```python
x_test.shape
```

```python
ct = make_column_transformer((OneHotEncoder(sparse=False),['Location']),remainder='passthrough')
```

```python
scaler=StandardScaler()
```

```python
model = LinearRegression(normalize=True)
```

```python
pipe = make_pipeline(ct,scaler,model)
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred_lr = pipe.predict(x_test)
```

```python
r2_score(y_test,y_pred_lr)
```
